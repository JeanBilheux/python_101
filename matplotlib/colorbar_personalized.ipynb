```python
import numpy as np
```

```python
LinL = [[0.5, 0.5, 0.5],
        [0.5, 0.5, 255.5],
        [255.5, 255.5, 0.5],
        [107, 212, 209],
        [107, 212, 209],
        [186, 75, 94],
        [155, 7, 19],
        [254, 3.5, 24],
        [184, 161, 139],
        [106, 241, 106]]
```

```python
LinL = np.array(LinL)
LinL = LinL/256
```

```python
print(LinL)
print(type(LinL))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
b3 = LinL[:,2] # value of blue at sample n
b2 = LinL[:,2] # value of blue at sample n
b1 = [0.0, 0.51, 0.58, 0.63, 0.69, 0.75, 0.79, 0.85, 0.93, 1]

# Setting up columns for tuples
g3 = LinL[:,1]
g2 = LinL[:,1]
g1 = b1

r3 = LinL[:,0]
r2 = LinL[:,0]
r1 = b1

# Creating tuples
R = zip(r1,r2,r3)
G = zip(g1,g2,g3)
B = zip(b1,b2,b3)

# Transposing
RGB = zip(R,G,B)
rgb = zip(*RGB)

# Creating dictionary
k = ['red', 'green', 'blue']
LinearL = dict(zip(k,rgb))
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap('matteo', LinearL)

plt.figure(figsize=(20,20))
plt.pcolor(rand(10,10),cmap=my_cmap)
plt.colorbar()
plt.show()
```
