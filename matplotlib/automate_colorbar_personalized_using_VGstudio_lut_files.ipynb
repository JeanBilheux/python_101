### Import LUP files created by VGstudio (export colorbar)

```python
%pylab inline
import numpy as np
import os
```

```python
dt_complex = np.dtype([('intensity', float),
                       ('r', float),
                       ('g', float),
                       ('b', float)])
```

```python
root_dir = '/Users/j35/Dropbox (ORNL)/Shawn'
```

```python
filename = 'jet.lut'
```

```python
full_filename = os.path.join(root_dir, filename)
print(full_filename)
print(os.path.isfile(full_filename))
```

```python
data = np.loadtxt(full_filename, dtype=dt_complex, skiprows=2)
```

```python
print(data)
```

```python
print(data['intensity'])
```

```python
data['intensity'][0] = 0
data['intensity'][-1] = 1
print(data['intensity'])
```

```python
LinL = zip(data['r'], data['g'], data['b'])
```

```python
print(LinL)
```

```python
new_LinL = np.array(LinL)
new_LinL = new_LinL / 256.
```

```python
print(new_LinL)
print(type(new_LinL))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
b3 = new_LinL[:,2] # value of blue at sample n
b2 = new_LinL[:,2] # value of blue at sample n
b1 = np.array(data['intensity'])

# Setting up columns for tuples
g3 = new_LinL[:,1]
g2 = new_LinL[:,1]
g1 = b1

r3 = new_LinL[:,0]
r2 = new_LinL[:,0]
r1 = b1

# Creating tuples
R = zip(r1,r2,r3)
G = zip(g1,g2,g3)
B = zip(b1,b2,b3)

# Transposing
RGB = zip(R,G,B)
rgb = zip(*RGB)

# Creating dictionary
k = ['red', 'green', 'blue']
LinearL = dict(zip(k,rgb))
```

```python
print(LinearL)
```

```python
my_cmap = matplotlib.colors.LinearSegmentedColormap('matteo', LinearL)

plt.figure(figsize=(20,20))
plt.pcolor(rand(10,10),cmap=my_cmap)
plt.colorbar()
plt.show()
```
