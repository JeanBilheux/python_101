```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(10)
y_axis = np.array([1,3,5,7,9,2,4,5,6,7])
error_axis = np.array([1,1,1,1,1,2,2,2,3,3])
```

```python
plt.figure(1)
plt.clf()
plt.errorbar(x_axis, y_axis, error_axis)
```

# with nan 

```python
x_axis = np.arange(10)
y_axis = np.array([1,3,5,7,9,4,4,5,6,7])
error_axis = np.array([1,1,1,np.nan,1,2,2,2,3,3])
```

```python
plt.figure(2)
plt.errorbar(x_axis, y_axis, error_axis)
```
