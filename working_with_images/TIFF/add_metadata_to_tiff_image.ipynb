# Learning how to load (PIL), manipulate data (numpy)  and save them back in tiff (16bits) (PIL)

Python Imaging Library (PIL) documentation can be found [here](http://effbot.org/imagingbook/pil-index.htm)

```python
from PIL import Image

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

import time
```

```python
filename = '/Users/j35/git/NeuNorm/tests/data/tif/sample/image003.tif'
```

# Loading image 

```python
image = Image.open(filename)
metadata_raw = dict(image.tag_v2)
image = np.array(image)
plt.figure()
plt.imshow(image)
plt.colorbar()
```

```python
print(metadata_raw)
```

## metadata to add 

```python
metadata = {1: 'this is value a',
           2: '5',
           3: '5'}
```

# Creating set of images 

```python
new_image = Image.fromarray(image)
new_image.save(filename, tiffinfo=metadata)
```
