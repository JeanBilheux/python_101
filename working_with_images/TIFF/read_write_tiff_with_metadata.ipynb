```python
from PIL import Image

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

import time
```

```python
filename = 'images/file_with_metadata_2018.tiff'
```

```python
image = Image.open(filename)
```

```python
plt.figure()
plt.imshow(image)
plt.colorbar()
```

```python
from PIL.ExifTags import TAGS
```

```python
info = image.tag.tags
```

```python
image.tag_v2.tagtype
```

```python
metadata = dict(image.tag_v2)
```

```python
type(metadata)
```

```python
metadata
```

```python
output_file = 'images/new_file_with_metadata_2018.tiff'
```

```python
new_image = np.array(image)
```

```python
plt.figure()
plt.imshow(new_image)
plt.colorbar()
```

```python
pillow_new_image = Image.fromarray(new_image)
pillow_new_image.save(output_file, tiffinfo=metadata)
```
