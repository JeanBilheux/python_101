```python
from NeuNorm.normalization import Normalization
import glob
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
list_of_images_to_load = glob.glob("../data/group_images_by_cycle/*.tif*")
print(f"nbr: {len(list_of_images_to_load)}")
```

```python
o_load = Normalization()
o_load.load(file=list_of_images_to_load, notebook=True)
```

# creating my own data

```python
bottom_right = np.zeros((40, 40))
bottom_right[0:20, 0:20] = 1

bottom_center = np.zeros((40, 40))
bottom_center[0:20, :] = 1

bottom_left = np.zeros((40, 40))
bottom_left[0:20, 20:] = 1
```

```python
middle_right = np.zeros((40, 40))
middle_right[:, 0:20] = 1

middle_center = np.ones((40, 40))

middle_left = np.zeros((40, 40))
middle_left[:, 20:] = 1
```

```python
top_right = np.zeros((40, 40))
top_right[20: , 0:20] = 1

top_center = np.zeros((40, 40))
top_center[20: , :] = 1

top_left = np.zeros((40, 40))
top_left[20: , 20:] = 1
```

```python
fig, axs = plt.subplots(3,3)
axs[2, 0].imshow(bottom_left)
axs[2, 1].imshow(bottom_center)
axs[2, 2].imshow(bottom_right)

axs[1, 0].imshow(middle_left)
axs[1, 1].imshow(middle_center)
axs[1, 2].imshow(middle_right)

axs[0, 0].imshow(top_left)
axs[0, 1].imshow(top_center)
axs[0, 2].imshow(top_right)

plt.tight_layout()
```

# saving the metadata 

```python
metadata = o_load.data['sample']['metadata']
```

# creating images with those metadata 

```python
o_new_images = Normalization()
o_new_images.load(data=[bottom_right,
                       bottom_center,
                       bottom_left,
                       middle_center,
                       top_left,
                       top_center,
                       top_right,
                       middle_right,
                       middle_center,
                       bottom_right,
                       bottom_center,
                       bottom_left,
                       middle_center,
                       top_left,
                       top_center,
                       top_right,
                       middle_right,
                       middle_center,
                       bottom_right,
                       bottom_center,
                       bottom_left,
                       middle_center,
                       top_left,
                       top_center,
                       top_right,
                       middle_right,
                       middle_center,
                       ],
                 notebook=True)
```

```python
o_new_images.data['sample']['metadata'] = metadata
```

# Export the new images 

```python
output_folder = "../data/group_images_by_cycle_with_my_own_data/"
o_new_images.export(data_type='sample', folder=output_folder)
```
