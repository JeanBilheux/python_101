```python
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

# User Input

```python
height = 5
width = 5
```

Sample

Initialization of image

```python
sample = np.ones((height, width))
```

```python
sample[:, 0] = 1
sample[:, 1] = 2
sample[:, 2] = 3
sample[:, 3] = 4
sample[:, 4] = 5       
sample[0, 0] = 10
sample[4, 4] = 10
```

```python
output_filename = '/Users/j35/Desktop/image_0001_roi_no_ob.tiff'
```

Let's check the image

```python
fig, ax = plt.subplots(1,1)
plt.imshow(sample)
```

```python
new_image = Image.fromarray(sample)
new_image.save(output_filename)
```
