```python
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

# User Input

Reference image

```python
# center of circle
x0 = 500
y0 = 600

# size of image
width = 1000
height = 1000
```

```python
# make array of distance of each pixel from center of circle
```

```python
x = np.arange(width) + 0.5
y = np.arange(height) + 0.5
```

```python
xv, yv = np.meshgrid(x, y)
distances_power = np.sqrt((np.power(yv - y0, 2) + np.power(xv - x0, 2)))
```

```python
np.shape(distances_power)
```

```python
fig, ax = plt.subplots(1,1)
plt.imshow(distances_power)
plt.colorbar()
```

export image

```python
new_image = Image.fromarray(distances_power)
output_file_name = "/Users/j35/Desktop/circle_profile.tiff"
new_image.save(output_file_name)
```
