```python
# init images
image1 = np.ones([10,10])
image1[0:5, 5:] = 5

image2 = np.ones([10,10])
image2[0:5, 0:5] = 10
image2[5:, 5:] = 5

# big images
big_array1 = np.zeros((10,15))
big_array2 = np.zeros((10,15)) 
final_array = np.zeros((10,15))

big_array1[0:10, 0:10] = image1
big_array2[0:10, 5:15] = image2

# overlap
final_array = np.zeros((10,15))
where_big_array1_has_value_only = np.where((big_array1 != 0) & (big_array2 == 0))
where_big_array2_has_value_only = np.where((big_array2 != 0) & (big_array1 == 0))
where_big_array1_and_array2_has_value = np.where((big_array1 !=0) & (big_array2 != 0))


final_array[where_big_array1_has_value_only] = big_array1[where_big_array1_has_value_only]
final_array[where_big_array2_has_value_only] = big_array2[where_big_array2_has_value_only]
final_array[where_big_array1_and_array2_has_value] = (big_array1[where_big_array1_and_array2_has_value] +
                                                     big_array2[where_big_array1_and_array2_has_value])/2

fig, axes = plt.subplots(1, 3)
axes[0].imshow(big_array1)
axes[1].imshow(big_array2)
axes[2].imshow(final_array)

```
