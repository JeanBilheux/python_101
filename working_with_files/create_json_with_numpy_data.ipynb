```python
import numpy as np
import json
```

```python
data = np.array([1, 2, 3, 4, 5])
metadata = {
    "description": "Sample numpy array",
    "shape": data.shape,}
```

```python
with open('data_with_metadata.json', 'w') as f:
    json.dump({
        "data": data.tolist(),
        "metadata": metadata
    }, f, indent=4)
```
