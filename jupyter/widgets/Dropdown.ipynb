```python
from ipywidgets import widgets
from IPython.display import display
```

### Example 1

```python
list_items = ['item1', 'item2', 'item3', 'item4', '']
result = widgets.Dropdown(options = list_items,
                          value = '',
                          allow_none = False,
                          sync = True, 
                          help = """This is the help""",
                         description='Choose your item')
display(result)
```

```python
print("value selected: %s " %result.value)
```

### Example 2

```python
result_2 = widgets.Dropdown(
    options={'One': 1, 'Two': 2, 'Three': 3},
    value=2,
    description='Number:',
)
display(result_2)
```

```python
print("value of result_2 selected is %s" %result_2.value)
```
