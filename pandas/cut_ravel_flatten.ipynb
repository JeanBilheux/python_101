```python
import pandas as pd
from itertools import product
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_df = pd.DataFrame([dict(x=x, y=y) for x,y in product(range(5), range(5))])
base_df['intensity'] = np.random.uniform(0, 1, 25)
base_df['transparency'] = np.random.uniform(0, 1, 25)
base_df.head()
```

# cut 

```python
base_df['x_cut'] = pd.cut(base_df['x'], 5)
base_df['y_cut'] = pd.cut(base_df['y'], 5)
```

# ravel (or flatten)

```python
fig, m_axes = plt.subplots(5, 5, figsize=(15, 15))
print(f"{np.shape(m_axes)}")
m_axes = m_axes.ravel()
print(f"{np.shape(m_axes)}")
for ((g_x, g_y), c_rows), c_ax in zip(base_df.sort_values(['x', 'y']).groupby(['x_cut', 'y_cut']), m_axes):
    c_ax.plot(c_rows['intensity'], c_rows['transparency'], 'r.')
```
