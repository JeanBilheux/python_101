This notebook will create a matrix, then using a given relative origin within the image, will calculate the angle of every single pixel relative to that pixel position

```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
matrix_height = 20
matrix_width = 20

x = np.arange(matrix_width)+ 0.5
y = np.arange(matrix_height)+ 0.5

yv, xv = np.meshgrid(y, x)

full_final_angles_matrix = np.zeros((matrix_height, matrix_width))
```

```python
relative_origin = np.array([5, 10])
```

## Bottom right corner

```python
right_bottom_corner_width = np.int(matrix_width - relative_origin[1])
right_bottom_corner_height = np.int(matrix_height- relative_origin[0])

x_right_bottom = np.arange(right_bottom_corner_width)+ relative_origin[1]
y_right_bottom = np.arange(right_bottom_corner_height)+ relative_origin[0]

xv_right_bottom, yv_right_bottom = np.meshgrid(x_right_bottom, y_right_bottom)

angles_right_bottom = 90 + np.rad2deg(np.arctan(((yv_right_bottom-relative_origin[0])) / 
                               (xv_right_bottom - relative_origin[1])))

plt.figure("bottom right")
plt.imshow(angles_right_bottom)

print(f"x_axis is {x_right_bottom}")
print(f"y_axis is {y_right_bottom}")
```

```python
full_final_angles_matrix[y_right_bottom[0]:, x_right_bottom[0]:] = angles_right_bottom
```

## Top right corner 

```python
right_top_corner_width = np.int(matrix_width - relative_origin[1])
right_top_corner_height = np.int(relative_origin[0])

x_right_top = np.arange(right_top_corner_width) + relative_origin[1]
y_right_top = np.arange(right_top_corner_height)

xv_right_top, yv_right_top = np.meshgrid(x_right_top, y_right_top)

angles_right_top = 90 + np.rad2deg(np.arctan((yv_right_top - relative_origin[0]) /
                                        (xv_right_top - relative_origin[1])))


plt.figure("top right")
plt.imshow(angles_right_top)

print(f"x_axis is {x_right_top}")
print(f"y_axis is {y_right_top}")
```

```python
full_final_angles_matrix[y_right_top[0]:y_right_top[0]+len(y_right_top), 
                         x_right_top[0]:x_right_top[0]+len(x_right_top)] = angles_right_top
```

## top left corner 

```python
left_top_corner_width = np.int(relative_origin[1])
left_top_corner_height = np.int(relative_origin[0])

x_left_top = np.arange(left_top_corner_width)
y_left_top = np.arange(relative_origin[0])

xv_left_top, yv_left_top = np.meshgrid(x_left_top, y_left_top)

angles_left_top = 270 + np.rad2deg(np.arctan((yv_left_top - relative_origin[0]) /
                                             (xv_left_top - relative_origin[1])))


plt.figure("top left")
plt.imshow(angles_left_top)

print(f"x_axis is {x_left_top}")
print(f"y_axis is {y_left_top}")
```

```python
full_final_angles_matrix[y_left_top[0]:y_left_top[0]+len(y_left_top), 
                         x_left_top[0]:x_left_top[0]+len(x_left_top)] = angles_left_top
```

## bottom left corner 

```python
left_bottom_corner_width = np.int(relative_origin[1])
left_bottom_corner_height = np.int(matrix_height- relative_origin[0])

x_left_bottom = np.arange(left_bottom_corner_width)
y_left_bottom = np.arange(left_bottom_corner_height) + relative_origin[0]

xv_left_bottom, yv_left_bottom = np.meshgrid(x_left_bottom, y_left_bottom)

angles_left_bottom = 270 + np.rad2deg(np.arctan((yv_left_bottom - relative_origin[0]) /
                                          (xv_left_bottom - relative_origin[1])))


plt.figure("bottom left")
plt.imshow(angles_left_bottom)

print(f"x_axis is {x_left_bottom}")
print(f"y_axis is {y_left_bottom}")
```

```python
full_final_angles_matrix[y_left_bottom[0]:y_left_bottom[0]+len(y_left_bottom), 
                         x_left_bottom[0]:x_left_bottom[0]+len(x_left_bottom)] = angles_left_bottom
```

## final result 

```python
plt.figure("final result")
plt.imshow(full_final_angles_matrix)
```
