```python
import matplotlib.pyplot as plt

import pyfits
import numpy as np

%matplotlib inline
```

```python
def read_single_fits(file):
    hdu_list = pyfits.open(file)  # fits
    hdu = hdu_list[0]
    _image = hdu.data
    _image = np.asarray(_image)
    hdu_list.close()
    return _image
```

```python
file_name = 'demo_file/Result of COL_SECTION0133.fits'
```

```python
data = read_single_fits(file_name)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(data)
```

```python
where_nan = np.isnan(data)
```

```python
data[where_nan] = 0
```

```python
plt.figure(figsize=(20,20))
plt.imshow(data)
```

```python
def make_fits(data=[], filename=''):
    hdu = pyfits.PrimaryHDU(data)
    hdulist = pyfits.HDUList([hdu])
    hdulist.writeto(filename)
    hdulist.close()
```

```python
output_file_name = 'demo_file/clean_data_set.fits'
make_fits(data=data, filename=output_file_name)
```
