How to label region of pixels and remove them from an image

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook

from scipy import ndimage
# https://docs.scipy.org/doc/scipy-0.16.0/reference/generated/scipy.ndimage.measurements.label.html
```

Define dummy data array with 3 groups of pixels
 - 2 groups (#1 and #2) at intensity of 5
 - 1 group (#3) at intensity of 2
 
Goal is to remove the group at intensity of 1

```python
data = np.zeros((20,20))

#group #1
data[3:4, 3:4] = 5

#group #2
data[10:13, 10:13] = 5

#group #3
data[15:18, 15:18] = 2

plt.imshow(data, cmap='viridis', interpolation=None)
```

```python
labeled_array, num_features = ndimage.measurements.label(data)
```

```python
group1 = (labeled_array == 1)
group2 = (labeled_array == 2)
group3 = (labeled_array == 3)

plt.subplots(1,3)

plt.subplot(131)
plt.imshow(group1, cmap='viridis')
plt.title("group1")
plt.subplot(132)
plt.imshow(group2, cmap='viridis')
plt.title("group2")
plt.subplot(133)
plt.imshow(group3, cmap='viridis')
plt.title("group3")
```

Let's remove group #3

```python
data[group3] = 0

plt.figure()
plt.imshow(data, cmap='viridis')
```
