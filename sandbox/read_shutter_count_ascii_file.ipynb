```python
import pandas as pd
import os
```

```python
input_file = "data/Run_7816_ShutterCount.txt"
assert os.path.exists(input_file), f"File not found: {input_file}"
```

```python
data = pd.read_csv(input_file, delim_whitespace=True, header=None, names=["ShutterCount"])
```

```python
data
```

```python
data = data[data["ShutterCount"] != 0].reset_index(drop=True)
```

```python
data["ShutterCount"].tolist()
```

```python
# keep only the shutter count until first 0
shutter_counts = data["ShutterCount"].tolist()
if 0 in shutter_counts:
    first_zero_index = shutter_counts.index(0)
    shutter_counts = shutter_counts[:first_zero_index]
```
