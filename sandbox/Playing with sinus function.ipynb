```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = np.arange(500)*0.01
```

```python
def sin(x, a, m, p, b):
    return a * np.sin(m*x+p) + b
```

```python
plt.figure(0)
plt.plot(x, sin(x, a=1, m=1, p=2, b=1), label='1,1,10,1')
plt.plot(x, sin(x, a=1, m=1, p=1, b=1), label='1,1,1,1')
plt.legend()
```
